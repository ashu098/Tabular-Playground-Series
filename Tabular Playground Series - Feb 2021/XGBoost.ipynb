```python
# importing libraries
import numpy as np
import pandas as pd

# checking all directories in our datasets
import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
train_path = '../input/tabular-playground-series-feb-2021/train.csv'
test_path = '../input/tabular-playground-series-feb-2021/test.csv'

# reading train.csv
train = pd.read_csv(train_path)
train.head()
```

```python
# rows and column in train folder
train.shape
```

```python
# reading test.csv
test = pd.read_csv(test_path)
test.head()
```

```python
test.shape
```

```python
extra = []
for i in train.columns:
    if i in test.columns:
        print(i,end=' ')
    else:
        extra.append(i)

print('\n\nextra column in train')
print(extra)
```

```python
target = train['target']
train.drop(['target'],axis = 1,inplace = True)

print(target)
```

```python
# as we don't have any null values in this dataset
train.isnull().values.any()
```

```python
for i in train.columns:
    if train[i].dtype == 'object':
        print(i,train[i].unique())
```

```python
# as we have columns of object type so convert them using label encoding

from sklearn.preprocessing import LabelEncoder

for i in train.columns:
    if train[i].dtype == 'object':
        labelencoder = LabelEncoder()
        
        labelencoder.fit(list(train[i].values))
        train[i] = labelencoder.transform(train[i].values)
        print(i,train[i].unique())
```

```python
for i in test.columns:
    if test[i].dtype == 'object':
        print(i,test[i].unique())
```

```python
# as we have columns of object type so convert them using label encoding

for i in test.columns:
    if test[i].dtype == 'object':
        labelencoder = LabelEncoder()
        
        labelencoder.fit(list(test[i].values))
        test[i] = labelencoder.transform(test[i].values)
        print(i,test[i].unique())
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
# read submission file
submission = pd.read_csv('../input/tabular-playground-series-feb-2021/sample_submission.csv',index_col='id')
```

```python
from sklearn.model_selection import train_test_split

X_train,X_valid,y_train,y_valid = train_test_split(train,target,
                                                   train_size=0.8,
                                                   test_size=0.2,
                                                   random_state=42)
```

```python
from sklearn.metrics import mean_absolute_error
from xgboost import XGBRegressor

# initalise xgb regressor
model = XGBRegressor(n_estimators = 100,learning_rate=0.01)

# train the model
model.fit(X_train,y_train,early_stopping_rounds=5,
               eval_set = [(X_valid,y_valid)],
              verbose=False)

#predict test dataset
submission['target'] = model.predict(test)
submission.to_csv('random_forest.csv')
```
